```python
import os
import dotenv
import sqlalchemy
import pandas as pd

# 1. Carica il file pswgym.env
dotenv.load_dotenv(dotenv_path="pswgym.env", override=True)

# 2. Leggi i dati usando le chiavi ESATTE del tuo file
user = os.getenv("user")
pw = os.getenv("pw")
host = os.getenv("host")
db = os.getenv("db")

#costruisco la stringa di connessione
connection_string = f"mysql+pymysql://{user}:{pw}@{host}/{db}"
#token di connessione (singleton) 
db_engine = sqlalchemy.create_engine(connection_string)


```

```python
query = "SELECT * FROM sessioni_allenamento"
#uso pandas e gli passo l sql da lanciare nel token di connessione
sessioni_allenamento = pd.read_sql(query, db_engine)

sessioni_allenamento 
```
